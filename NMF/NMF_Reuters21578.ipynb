```python
def tokenize(text, lower=True):
    
    regex1 = re.compile('[^a-zA-Z]')
    if lower:
        corpus = regex1.sub(' ', text.lower())
        words = [stemmer.stem(word) for word in word_tokenize(corpus) if word not in stopwords]
    else:
        corpus = regex1.sub(' ', text.lower())
        words = [stemmer.stem(word) for word in word_tokenize(corpus) if word not in stopwords]
    
    return words 
```

```python
def topic_words(clf, vectorizer, n_words = 15):
    vocabulary = np.array(vectorizer.get_feature_names())
    X = clf.components_
    top_words = lambda t: [vocabulary[i] for i in np.argsort(t)[:-n_words-1:-1]]
    topic_words = ([top_words(t) for t in X])
    return [' '.join(t) for t in topic_words]

topic_words(nmf0, bow0)
```
